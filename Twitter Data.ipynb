```python
# Import dependencies
import pandas as pd
import datetime as dt 
```

```python
# Define filed and convert to dataframes
file_1= "Data/covidvaccine.csv"
file_2= "Data/vaccination_all_tweets.csv"

df_1= pd.read_csv(file_1, parse_dates=False)
df_2= pd.read_csv(file_2, parse_dates=False)
```

```python
df_1.head()
```

```python
# Check data types
df_1.dtypes
```

```python
# Drop unnecessary columns

df_1_final= df_1.drop(columns= ["user_description", "user_created", "user_friends", "user_followers","user_favourites","user_verified", "source"])
df_1_final.head
```

```python
# Check for null values in text column
df_1_final.text.isnull().sum
```

```python
# Repeat for second dataframe
df_2.head()
```

```python
# Check data types
df_2.dtypes
```

```python
# Drop unnecessary columns

df_2_final= df_2.drop(columns= ["id","user_description", "user_created", "user_friends", "user_followers","user_favourites","user_verified", "source","retweets", "favorites"])
df_2_final.head
```

```python
# Concatenate dataframes
final_df= pd.concat([df_1_final, df_2_final], ignore_index=True)
final_df.head()
```

```python
final_df.dtypes
```

```python
# Drop the duplicate text values
final_df= final_df.drop_duplicates(subset="text")
final_df
```

```python
blank_location= final_df.user_location.isnull().sum()
blank_location
```

```python
# Drop rows where user_location is blank
vaccine_tweets_df = final_df[final_df['user_location'].notna()]
vaccine_tweets_df
```

```python
vaccine_tweets_df.dtypes
```

```python
vaccine_tweets_df["date"]= pd.to_datetime(vaccine_tweets_df["date"], errors= "coerce")                                
```
